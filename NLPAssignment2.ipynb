```
import re
from nltk.corpus import wordnet
R_patterns = [
   (r'won\'t', 'will not'),
   (r'can\'t', 'cannot'),
   (r'i\'m', 'i am'),
   (r'(\w+)\'ll', '\g<1> will'),
   (r'(\w+)n\'t', '\g<1> not'),
   (r'(\w+)\'ve', '\g<1> have'),
   (r'(\w+)\'s', '\g<1> is'),
   (r'(\w+)\'re', '\g<1> are'),
]
```

Now we need to convert words

```
pattern = [(re.compile(regex), repl) for (regex, repl) in R_patterns ]
```

```
s = raw_text
len(s)
```

```
#s = raw_text
for (pat, repl) in pattern:
    s = re.sub(pat, repl, s)
```

```
def replaceWords(s):
    pattern = [(re.compile(regex), repl) for (regex, repl) in R_patterns ]
    for (pat, repl) in pattern:
        s = re.sub(pat, repl, s)

    return s
```

```
def repeatingWord(word):
    repeat_regexp = re.compile(r'(\w*)(\w)\2(\w*)')
    repl = r'\1\2\3'
    if wordnet.synsets(word):
        return word
    repl_word = repeat_regexp.sub(repl, word)
    if repl_word != word :
        return repeatingWord(repl_word)
    else:
        return repl_word    
```

```
import csv

class syn_replacer(object):

    def __init__(self):
        self.word_map = {}

        length = filedf.shape[0]
        for i in range(length):
            self.word_map[filedf.iloc[i,0]] = filedf.iloc[i,1]

    
    def replace(self, word):
        return self.word_map.get(word, word)

    def replace_sent(self, sent):
        w_list = sent.split(" ")
        w_list = [self.word_map.get(w, w) for w in w_list]
        return w_list
```

```
#synonym replacer's object
syn_rep = syn_replacer()
print(syn_rep.replace('ttyl'))
print(syn_rep.replace_sent('Today is my bday'))
print(syn_rep.replace_sent('lol hi, I will ttyl'))
```

Antonyms

```
class word_antonym_replacer(object):

  def replace(self, word, pos=None):
    
    antonyms = set()
    
    for syn in wordnet.synsets(word, pos=pos):
        for lemma in syn.lemmas():
          for antonym in lemma.antonyms():
            antonyms.add(antonym.name())
    
    
    if len(antonyms) == 1:
        return antonyms.pop()
    else:
        return None
  
  def replace_negations(self, sent):
    i, l = 0, len(sent)
    words = []
    
    while i < l:
        word = sent[i]
        
        if word == 'not' and i+1 < l:
          ant = self.replace(sent[i+1])
          
          if ant:
              words.append(ant)
              i += 2
              continue
        
        words.append(word)
        i += 1
    
    return words
```

```
rep_ant = word_antonym_replacer()

print(rep_ant.replace('ugly'))
print(rep_ant.replace_negations(('This laptop is not heavy').split(' ')))
print(rep_ant.replace_negations(('Pyrotechniques are not dangerous').split(' ')))
```
